```python
# Univariate Analysis of Stationary Time Series Employment

emp=UK[["Employment","Year"]]
```

```python
#Remove NaN
emp_na = emp.dropna(subset=['Employment'])
```

```python
print(emp_na)
```

```python
#1.2 Convertir emp_na en una serie temporal 
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
#1.2.1 Convertir años en serie temporal y transformar los años en un indice 
emp_na["Year"]=pd.to_datetime(emp_na["Year"],format="%Y") # convertimos los años en serie temporal
emp_na.set_index("Year",inplace=True)
#
#En Python, %Y es un especificador de formato utilizado en la función strptime() de la biblioteca datetime para analizar cadenas de fecha y hora en objetos datetime.
#Específicamente, %Y se utiliza para representar el año con cuatro dígitos. Por ejemplo, si tienes una cadena de fecha como "2024", %Y la interpretará como el año 2024.
```

```python
# 1.2.2 convertimos Employment emp_na en una serie temporal (ahora year es un indice)
emp_na_ts = emp_na['Employment']
```

```python
print(emp_na_ts)
```

```python
#1.3 Transformar la variable en logaritmos
log_emp_na_ts=np.log(emp_na_ts)
```

```python
# Print the result
print(log_emp_na_ts)
```

```python
import matplotlib.pyplot as plt

plt.plot(log_emp_na_ts.index, log_emp_na_ts.values)
plt.xlabel('Year')
plt.ylabel('Log Employment')
plt.title('Log Transformed Employment Over Time')
plt.xticks(log_emp_na_ts.index[::10], rotation=45)  # Ajustar los marcadores para mostrar cada 10 años
plt.tight_layout()
plt.show()
```

```python
acf_result = sm.tsa.acf(emp_na_ts, nlags=len(emp_na_ts)-1)
```

```python
# 2 Analysis
plt.stem(acf_result)
plt.title('Autocorrelaciones')
plt.xlabel('Retraso')
plt.ylabel('Autocorrelación')
plt.show()

# Vemos muchisima persistencia, esto podria tratarse de un random walk por lo que la serie no es estacionaria 
```

```python
pacf = sm.tsa.pacf(emp_na_ts)

# Plot PACF
plt.stem(range(len(pacf)), pacf)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function')
plt.show()

# De aqui puedo entedrer un AR(1), ninguno de los otros periodos parece significativo 
```

```python
# 3 Transformar la variable es estacionaria, aplicar differencias
d_emp_na = emp_na_ts.diff()

# Plot differenced series
plt.plot(d_emp_na.index, d_emp_na.values)
plt.xlabel('Year')
plt.ylabel('Differenced Employment')
plt.title('Differenced Employment Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#3.1 Generate ac function with d_emp_na
acf_result_d = sm.tsa.acf(d_emp_na, nlags=len(d_emp_na)-1)
```

```python
# 3.2 Analysis
plt.stem(acf_result_d)
plt.title('Autocorrelaciones')
plt.xlabel('Retraso')
plt.ylabel('Autocorrelación')
plt.show()
```

```python
acf_result = sm.tsa.acf(d_emp_na, nlags=len(d_emp_na))

# Grafica la función de autocorrelación
plt.stem(acf_result)
plt.title('Autocorrelations')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.ylim(-1, 1)
plt.show()
```

```python
pacf = sm.tsa.pacf(d_emp_na)

# Plot PACF
plt.stem(range(len(pacf)), pacf)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function')
plt.show()
```
